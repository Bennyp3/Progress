```python
# import the packages
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# let's create some random data to work with
x = np.array(range(0,100))
y = np.random.rand(100,1) # create a 100 by 1 array (like a list with 100 random numbers)
plt.plot(x,y)
```

#### Let's do some styling 

```python
# you can use the keyword 'marker' to emphasize each point with a specified marker

plt.plot(x,y, marker = '*') # other markers you may use are: 'o':Circle, '*':Star '.':Point ',':Pixel, +,P,s,D,d,p,H,v,^,<,>,1, etc 
# you can find more here: https://www.w3schools.com/python/matplotlib_markers.asp
```

```python
# you may combine the marker, line type and color in a format string
# marker line color

# let's try a red dashed plot with hexagonal markers

plt.plot(x,y, 'H--r')  

# lines: -,:,--,-.
# colors: r,g,b,c,m,y,k,w
```

```python
# change the marker size, markeredge color and marker background color 
# keyword arguments: markersize or ms, markeredgecolor or mec, markerfacecolor or mfc

plt.plot(x,y,marker = 'H', markersize = 15, markerfacecolor = 'blue', markeredgecolor = 'red')
# same as 
# plt.plot(x,y,marker = 'H', ms = 15, mfc = 'blue', mec = 'red')

#  140 supported color names and their hexadecimal values https://www.w3schools.com/colors/colors_names.asp
# plt.plot(x,y,marker = 'H', ms = 15, mfc = '#0000FF', mec = '#FF0000')
```

```python
# linestyle

# linestyle argument can be shortened to ls

# plt.plot(x,y, ls = '--')  
#linestyle symbols -,:,--,-.
# linestyle names: 'solid', 'dotted', 'dashed', 'dashdot',None  
# linewidth or lw (pt)  
plt.plot(x,y, ls=':', color = 'blue', lw = 2, ms = 5, marker = 'H', mfc = 'k', mec = 'green')
```

```python
# multiple lines
# add multiple lines by calling more plt.plot() functions:
# let's create some more random data
y2 = np.random.rand(100,1)
plt.plot(x,y, ls=':', color = 'blue', lw = 2, ms = 5, marker = 'H', mfc = 'k', mec = 'green', label = 'line1')
plt.plot(y2, c = 'red', label = 'line2')
plt.legend(loc = 9) 
# other locations: 'best':0 , 'upper right':1, 'upper left':2, 'lower left':3, 'lower right':4
# 'right':5 'center left':6 'center right':7 'lower center':8 'upper center':9 'center':10
#  https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.legend.html
```

```python
# label the plot
# xlabel, ylabel, title
plt.plot(y2, c = 'red')
plt.xlabel('x')
plt.ylabel('random noise')
plt.title('INFO2000 plotting')
plt.show()
```

```python
# set fonts of the labels using a dictionary font
# set location of title using loc: could be left, right or center
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("info2000", fontdict=font1, loc = 'left')
plt.xlabel("x", fontdict=font2)
plt.ylabel('random noise', fontdict=font2)
plt.plot(y2)
plt.show()
```

```python
# grid lines
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title("info2000", fontdict=font1, loc = 'left')
plt.xlabel("x", fontdict=font2)
plt.ylabel('random noise', fontdict=font2)
plt.plot(y2)
plt.grid()
# plt.grid(axis = 'x')
# plt.grid(axis='y')
# plt.grid(color = 'red', ls = '-.',lw  = 0.5)
plt.show()
```

### Subplots

```python
# access the subplots explicitly
fig,ax = plt.subplots(2,2)
ax[0][0].plot(y2)
ax[0][1].plot(y, lw =0.5, ls = ':', c = 'red')
ax[0][1].set_title('part1')
ax[0][1].set_xlabel('x')
ax[0][1].set_ylabel('y')
```

```python

plt.subplot(2,2,1)
plt.plot(y, c = 'k', label = 'black')
plt.title('black')
plt.legend(loc = 1)
plt.subplot(2,2,2)
plt.plot(x,y2, c = 'g', label = 'green')
plt.title('green')
plt.legend(loc = 1)
plt.subplot(2,2,3)
plt.plot(x,y2, c= 'y', label = 'yellow')
plt.title('yellow')
plt.legend(loc = 1)
plt.subplot(2,2,4)
plt.plot(x,y2, c = 'r', label = 'red')
plt.title('red')
plt.legend(loc = 1)
plt.suptitle("ALL")
plt.legend(loc = 1)
plt.show()
```
